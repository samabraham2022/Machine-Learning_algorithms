```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df.head() # to view the first few rows
df.info() # to get the information about the dataset
df.describe() # to get the statistical summary of the dataset
```

```python
df['Item_Weight'].fillna(df['Item_Weight'].mean(), inplace=True)
df['Item_Fat_Content'].replace({'Low Fat': 0, 'Regular': 1, 'LF': 0, 'reg': 1, 'low fat': 0}, inplace=True)
df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0], inplace=True)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Item_Type'], df['Item_MRP'], alpha=0.5,color='red')
plt.xticks(rotation=45, ha='right') 

plt.xlabel('Item Type')
plt.ylabel('Item MRP')
plt.title('Item Type vs Item MRP')
plt.show()

plt.figure(figsize=(10,6))
plt.bar(df['Outlet_Location_Type'], df['Item_Outlet_Sales'],color='red')
plt.xlabel('Outlet Location Type')
plt.ylabel('Item Outlet Sales')
plt.title('Outlet Location Type vs Item Outlet Sales')
plt.show()
```

```python
item_type_sales = df.groupby(['Item_Type'])['Item_Outlet_Sales'].mean()
top_10 = item_type_sales.sort_values(ascending=False).head(10)
colors = ['red'] * 10 
colors[list(top_10.index).index(top_10.index[0])] = 'green' 
plt.figure(figsize=(10,6))
plt.bar(top_10.index, top_10.values, color=colors) 
plt.xticks(rotation=45, ha='right')
plt.xlabel('Item Type')
plt.ylabel('Mean Item Outlet Sales')
plt.title('Top 10 Products with the Highest Impact on Sales')
plt.show()
```

The highest impact making Food is highlighted in green
